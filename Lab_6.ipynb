```python
word_freq_dict = collections.defaultdict(int)

for word in words:
    word_freq_dict[' '.join(word) + ' </w>'] += 1
```

```python
def get_pairs(word_freq_dict):
    pairs = collections.defaultdict(int)
    for word, freq in word_freq_dict.items():
        chars = word.split()
        for i in range(len(chars)-1):
            pairs[chars[i], chars[i+1]] += freq
    return pairs

def merge_byte_pairs(best_pair, word_freq_dict):
    merged_dict = {}
    bigram = re.escape(' '.join(best_pair))
    p = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
    for word in word_freq_dict:
        # print(word)
        w_out = p.sub(''.join(best_pair), word)
        merged_dict[w_out] = word_freq_dict[word]
    return merged_dict

def get_subword_tokens(word_freq_dict):
    char_freq_dict = collections.defaultdict(int)
    for word, freq in word_freq_dict.items():
        chars = word.split()
        for char in chars:
            char_freq_dict[char] += freq
    return char_freq_dict

rules = []

for i in range(100):
    pairs = get_pairs(word_freq_dict)
    best_pair = max(pairs, key=pairs.get)
    rules.append(best_pair)
    print("best_pair --->", best_pair)
    print(f"Iteration {i}: ")
    word_freq_dict = merge_byte_pairs(best_pair, word_freq_dict)
    subword_tokens = get_subword_tokens(word_freq_dict)
    print(subword_tokens)
    # print(len(subword_tokens))
    print("----------------------------------------------------------------------------------------------------------------------------------------")
```

```python
test_sentence_tokens = []
for word in test_words:
  for char in word:
    test_sentence_tokens.append(char)
  test_sentence_tokens.append('</w>')
test_sentence_tokens
```

```python
modified_test_sentence = ' '.join(test_sentence_tokens)
modified_test_sentence
```

```python
def replace_bigrams(tokens, rules):
    cnt = 0
    for bigram in rules:
        cnt += 1
        bigram_str = ''.join(bigram)
        i = 0
        while i < len(tokens) - 1:
            if tokens[i] == bigram[0] and tokens[i + 1] == bigram[1]:
                tokens[i] = bigram_str
                del tokens[i + 1]
            else:
                i += 1
        print(f"Rule {cnt}: {bigram[0]} + {bigram[1]} = {bigram_str}")
        print(f"Modified Sentence:", ' '.join(tokens))
        print("----------------------------------------------------------------------------------------------------------------------------------------")
```

```python
temp = test_sentence_tokens
replace_bigrams(temp, rules)
```
