# Preprocessing documents

```python
def preprocessDoc(url):
  def getHTMLdocument(url):
      response = requests.get(url)
      return response.text
  html_document = getHTMLdocument(url)
  # print(html_document)
  soup = BeautifulSoup(html_document, 'html.parser')
  # print(soup.prettify())
  org_text = soup.get_text()
  # print(org_text)

  para_tokens = org_text.split('\n')
  para_tokens = [t for t in para_tokens if t]
  # para_tokens


  removedPunctPara = []

  for p in para_tokens:
    temp = p.translate(str.maketrans('', '', string.punctuation))
    removedPunctPara.append(temp)


  sentence_tokens = []
  for p in removedPunctPara:
    sentences = nltk.sent_tokenize(p)
    for s in sentences:
      sentence_tokens.append(s.strip())


  words_tokens = []

  for s in sentence_tokens:
    words = nltk.word_tokenize(s)
    for w in words:
      words_tokens.append(w)

  stop_words = set(stopwords.words('english'))
  words_tokens = [w for w in words_tokens if not w.lower() in stop_words]

  stemmer_ps = PorterStemmer()
  stemmed_words = [stemmer_ps.stem(word) for word in words_tokens]

  lemmatizer = WordNetLemmatizer()
  lemmatized_words = [lemmatizer.lemmatize(word) for word in stemmed_words]
  print(lemmatized_words)

  return lemmatized_words
```

```python
corpus = []
for url in sports_urls:
  doc = preprocessDoc(url)
  corpus.append([doc, 0])

for url in education_urls:
  doc = preprocessDoc(url)
  corpus.append([doc, 1])
```

```python
for row in corpus:
  print(row)
```

# Unigram Count Matrix

```python
output = []
for url in sports_urls:
  output.append(0)

for url in education_urls:
  output.append(1)

output
```

```python
unique_words = set()

for doc in corpus:
  for token in doc[0]:
      unique_words.add(token)
```

```python
len(unique_words)
```

```python
df = pd.DataFrame(0, index=range(len(corpus)), columns=list(unique_words))
for i, doc in enumerate(corpus):
    for word in doc[0]:
        df.at[i, word] += 1

df['output'] = output
```

# Bigram Probabilty Matrix

```python
bigrams = set()
for doc in corpus:
  for i in range(0, len(doc[0]) - 1):
    curr = doc[0][i]
    next = doc[0][i + 1]
    new = curr + ' ' + next
    bigrams.add(new)
len(bigrams)
```

```python
bi_df = pd.DataFrame(0, index=range(len(corpus)), columns=list(bigrams))
for i, doc in enumerate(corpus):
    for j in range(len(doc[0]) - 1):
        bigram = doc[0][j] + ' ' + doc[0][j + 1]
        first_word = doc[0][j]
        # Get the count of the first word in the unigram count matrix
        first_word_count = df.loc[i, first_word] if first_word in df.columns else 0
        if first_word_count > 0:  # Check to avoid division by zero
            # Divide the bigram count by the first word's count to get the probability
            bi_df.at[i, bigram] = bi_df.at[i, bigram] / first_word_count
```

```python
bi_df['output'] = output

bi_df
```

# TF - IDF Matrix

```python
tf_mat = df.copy()
tf_mat = tf_mat.div(tf_mat.sum(axis=1), axis=0)
tf_mat
```

```python
count_greater_than_zero = df.apply(lambda col: (col > 0).sum())
# print(count_greater_than_zero)
count_greater_than_zero = len(corpus) / count_greater_than_zero
# count_greater_than_zero
count_greater_than_zero = count_greater_than_zero.apply(lambda x: math.log2(x))
count_greater_than_zero
```

```python
tfidf_mat = tf_mat.mul(count_greater_than_zero, axis=1)
tfidf_mat['output'] = output
tfidf_mat
```

# Classification

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = df.drop('output', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, df['output'], test_size=0.2, random_state=42)

model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
X = bi_df.drop('output', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, bi_df["output"], test_size=0.2, random_state=42)

model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
X = tfidf_mat.drop('output', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, tfidf_mat["output"], test_size=0.2, random_state=42)

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy_score(y_test, y_pred)
```
